```python
# Setup
import sys
from pathlib import Path
import json
import logging

project_root = Path.cwd().parent
sys.path.insert(0, str(project_root / "src"))

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

print(f"✓ Project Root: {project_root}")
```

## Stage 1: Load Configuration

```python
from config import Config

print("\n" + "="*60)
print("CONFIGURATION")
print("="*60)

Config.validate()

print(f"\nQdrant URL: {Config.QDRANT_URL}")
print(f"Collection: {Config.QDRANT_COLLECTION_NAME}")
print(f"Data Folder: {Config.DATA_FOLDER}")
print(f"OpenAI Model: {Config.OPENAI_MODEL}")
```

## Stage 2: Load Documents

```python
from multi_file_loader import MultiFileLoader

print("\n" + "="*60)
print("STAGE 2: LOADING DOCUMENTS")
print("="*60 + "\n")

loader = MultiFileLoader()

# Load all documents from data folder
documents = loader.load_all_documents(
    folder_path=Config.DATA_FOLDER,
    recursive=True
)

print(f"\n✅ Loaded {len(documents)} documents")
print(f"   Total elements: {sum(doc.get('element_count', 0) for doc in documents)}")
```

## Stage 3: Link Related Content

```python
from context_preserver import HierarchicalContextPreserver

if len(documents) == 0:
    print("⚠️  No documents to process. Add files to data folder.")
else:
    print("\n" + "="*60)
    print("STAGE 3: LINKING CONTENT")
    print("="*60 + "\n")
    
    preserver = HierarchicalContextPreserver()
    enriched_elements = preserver.process_documents(documents)
    
    print(f"\n✅ Created {len(enriched_elements)} enriched elements")
    
    # Show statistics
    stats = preserver.get_statistics()
    print(f"\nStatistics:")
    print(f"  Files processed: {stats['total_files']}")
    print(f"  Total elements: {stats['total_elements']}")
    print(f"  Relationships: {stats['total_relationships']}")
```

## Stage 4: Summarize Content

```python
from summarizer import IntelligentSummarizer

if len(documents) == 0:
    print("⚠️  No documents to summarize. Add files to data folder.")
else:
    print("\n" + "="*60)
    print("STAGE 4: SUMMARIZING CONTENT")
    print("="*60 + "\n")
    print("⚠️  This will make API calls to OpenAI (costs money)")
    print(f"   Estimated cost for {len(enriched_elements)} elements: $1-5")
    
    proceed = input("\nProceed with summarization? (yes/no): ")
    
    if proceed.lower() in ['yes', 'y']:
        summarizer = IntelligentSummarizer(
            openai_api_key=Config.OPENAI_API_KEY,
            model="gpt-4o-mini"  # Cost-effective model
        )
        
        # Process first 10 elements as demo
        print(f"\n📝 Processing first 10 elements (demo)...")
        processed_elements = summarizer.process_elements(enriched_elements[:10])
        
        print(f"\n✅ Summarized {len(processed_elements)} elements")
        
        # Show statistics
        stats = summarizer.get_statistics()
        print(f"\nStatistics:")
        print(f"  Text summarized: {stats['text_summarized']}")
        print(f"  Images described: {stats['images_described']}")
        print(f"  Code analyzed: {stats['code_processed']}")
        print(f"  Tables summarized: {stats['tables_summarized']}")
        print(f"  Total API calls: {stats['total_api_calls']}")
        
        # Show example
        if len(processed_elements) > 0:
            print(f"\n" + "="*60)
            print("EXAMPLE ELEMENT (with summaries)")
            print("="*60)
            elem = processed_elements[0]
            print(f"\nElement ID: {elem['element_id']}")
            print(f"Type: {elem['type']}")
            print(f"File: {elem['filename']}")
            if 'summary_short' in elem:
                print(f"\nShort Summary: {elem['summary_short']}")
            if 'summary_medium' in elem:
                print(f"\nMedium Summary: {elem['summary_medium'][:200]}...")
    else:
        print("\n⏭️  Skipping summarization. Using elements without summaries.")
        processed_elements = enriched_elements
```

```python
# Save pipeline results for inspection
if len(documents) > 0 and 'enriched_elements' in locals():
    output = {
        'total_documents': len(documents),
        'total_elements': len(enriched_elements),
        'file_stats': loader.file_stats,
        'context_stats': preserver.get_statistics(),
        'sample_element': enriched_elements[0] if len(enriched_elements) > 0 else None
    }
    
    output_file = project_root / "pipeline_results.json"
    
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(output, f, indent=2, ensure_ascii=False, default=str)
    
    print(f"\n✓ Results saved to: {output_file}")
```
