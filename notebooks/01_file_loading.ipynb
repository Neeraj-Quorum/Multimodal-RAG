```python
# Setup
import sys
from pathlib import Path
import json

project_root = Path.cwd().parent
sys.path.insert(0, str(project_root / "src"))

print(f"✓ Project Root: {project_root}")
```

## Step 1: Initialize the Multi-File Loader

```python
from multi_file_loader import MultiFileLoader
import logging

# Setup logging to see detailed output
logging.basicConfig(level=logging.INFO, format='%(message)s')

# Initialize loader
loader = MultiFileLoader()
print("\n✅ Multi-File Loader ready!")
```

## Step 2: Test Individual File Types

```python
# List supported file types
print("Supported File Types:")
print("=" * 60)
for file_type, extensions in loader.SUPPORTED_EXTENSIONS.items():
    print(f"{file_type:10s}: {', '.join(extensions)}")
print("=" * 60)
```

### Test with Sample Files (if available)

```python
# Find sample files in your data folder
data_folder = project_root / "data"

if not data_folder.exists():
    print(f"⚠️  Data folder not found: {data_folder}")
    print(f"Creating it now...")
    data_folder.mkdir(parents=True, exist_ok=True)
    print(f"✓ Created: {data_folder}")
    print(f"\n💡 Please add your documents to: {data_folder}")
else:
    print(f"✓ Data folder exists: {data_folder}")
    
# List all files
all_files = list(data_folder.glob("**/*"))
file_list = [f for f in all_files if f.is_file()]

print(f"\nFound {len(file_list)} files in data folder")

if len(file_list) > 0:
    print("\nFirst 10 files:")
    for i, f in enumerate(file_list[:10], 1):
        print(f"  {i}. {f.name} ({f.suffix})")
else:
    print("\n⚠️  No files found. Add your documents to test loading.")
```

## Step 3: Load All Documents from Data Folder

```python
print("\n" + "="*60)
print("LOADING ALL DOCUMENTS")
print("="*60 + "\n")

# Load all documents recursively
documents = loader.load_all_documents(
    folder_path=data_folder,
    recursive=True  # Scan subfolders
)

print(f"\n✅ Loaded {len(documents)} documents successfully!")
```

## Step 4: Analyze Loaded Documents

```python
if len(documents) == 0:
    print("⚠️  No documents loaded. Add files to the data folder.")
    print(f"\n📁 Data folder: {data_folder}")
    print("\nSupported formats: PDF, DOCX, HTML, TXT, PNG, JPG, Code files")
else:
    print("\n" + "="*60)
    print("DOCUMENT ANALYSIS")
    print("="*60)
    
    # Count elements by type
    total_text_chunks = sum(len(doc.get('text_chunks', [])) for doc in documents)
    total_tables = sum(len(doc.get('tables', [])) for doc in documents)
    total_images = sum(len(doc.get('images', [])) for doc in documents)
    total_elements = sum(doc.get('element_count', 0) for doc in documents)
    
    print(f"\nTotal Documents:    {len(documents)}")
    print(f"Total Elements:     {total_elements}")
    print(f"  - Text Chunks:    {total_text_chunks}")
    print(f"  - Tables:         {total_tables}")
    print(f"  - Images:         {total_images}")
    print("\n" + "="*60)
```

## Step 5: Preview First Document

```python
if len(documents) > 0:
    print("\n" + "="*60)
    print("FIRST DOCUMENT PREVIEW")
    print("="*60 + "\n")
    
    doc = documents[0]
    
    print(f"File ID:       {doc.get('file_id')}")
    print(f"Filename:      {doc.get('filename')}")
    print(f"Type:          {doc.get('type')}")
    print(f"Element Count: {doc.get('element_count')}")
    print(f"Loaded At:     {doc.get('loaded_at')}")
    
    # Show first text chunk if available
    if len(doc.get('text_chunks', [])) > 0:
        first_chunk = doc['text_chunks'][0]
        print(f"\nFirst Text Chunk:")
        print(f"  Element ID: {first_chunk.get('element_id')}")
        print(f"  Content Preview: {first_chunk.get('content', '')[:200]}...")
    
    # Show structure
    print(f"\nDocument Structure:")
    print(json.dumps(
        {k: v for k, v in doc.items() if k not in ['text_chunks', 'tables', 'images', 'content']},
        indent=2
    ))
```

## Step 6: Save Loaded Documents (Optional)

```python
# Save to JSON for inspection
if len(documents) > 0:
    output_file = project_root / "loaded_documents.json"
    
    # Simplify for JSON (remove large binary data)
    simplified_docs = []
    for doc in documents:
        simplified = doc.copy()
        # Remove large base64 image data
        if 'image_data' in simplified:
            simplified['image_data'] = f"<base64 data, {len(simplified['image_data'])} chars>"
        if 'images' in simplified:
            for img in simplified['images']:
                if 'content' in img and len(img['content']) > 100:
                    img['content'] = f"<base64 data, {len(img['content'])} chars>"
        simplified_docs.append(simplified)
    
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(simplified_docs, f, indent=2, ensure_ascii=False)
    
    print(f"\n✓ Saved to: {output_file}")
    print(f"  You can inspect the structure manually")
```
